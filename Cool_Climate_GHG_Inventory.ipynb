```python
import pandas as pd
import numpy as np
```

# Importing the Datasets

```python
"""

Import pre-treated Agricultural data from USDA NASS into dataframe: https://quickstats.nass.usda.gov/results/15C3BC70-E0ED-3783-936B-6FE3805765FA
Manually created sub-labels for Agricultural Data to group livestock into: Crop, Cattle, Swine, Poultry, Other Livestock (Sheep, Goat, and Horse), Other Livestock* (all other than previous Other Livestock), Not Specified, amd some blank values for those that we couldn't identify.

"""
county_list = ['ALAMEDA','ALPINE','AMADOR','BUTTE','CALAVERAS','COLUSA','CONTRA COSTA','DEL NORTE','EL DORADO','FRESNO','GLENN','HUMBOLDT','IMPERIAL','INYO','KERN','KINGS','LAKE','LASSEN','LOS ANGELES','MADERA','MARIN','MARIPOSA','MENDOCINO','MERCED','MODOC','MONO','MONTEREY','NAPA','NEVADA','ORANGE','PLACER','PLUMAS','RIVERSIDE','SACRAMENTO','SAN BENITO','SAN BERNARDINO','SAN DIEGO','SAN FRANCISCO','SAN JOAQUIN','SAN LUIS OBISPO','SAN MATEO','SANTA BARBARA','SANTA CLARA','SANTA CRUZ','SHASTA','SIERRA','SISKIYOU','SOLANO','SONOMA','STANISLAUS','SUTTER','TEHAMA','TRINITY','TULARE','TUOLUMNE','VENTURA','YOLO','YUBA']
agriculture_census = pd.read_csv('/Agriculture Census Data (with label).csv')
```

```python
#import GHG inventory sets from California air Resources Board
emissions_2000_2021 = pd.read_csv('/GHG Emissions (2000-2021).csv')
inventory_2000_2021 = emissions_2000_2021
inventory_2000_2021['N2O'] = inventory_2000_2021['GHG'] == 'N2O'
inventory_2000_2021['CO2'] = inventory_2000_2021['GHG'] == 'CO2'
inventory_2000_2021['CH4'] = inventory_2000_2021['GHG'] == 'CH4'

for i in range(2000,2022):
  inventory_2000_2021[str(i)] = inventory_2000_2021[str(i)] * (inventory_2000_2021['N2O'] * 265 + inventory_2000_2021['CO2'] * 1 + inventory_2000_2021['CH4'] * 28)

inventory_2000_2021 = inventory_2000_2021.groupby(['Sublabel']).sum().drop(['Inventory Accountin','Main Sector', 'Sub Sector Level 1', 'Sub Sector Level 2', 'Sub Sector Level 3', 'Main Activity', 'Activity Subset', 'GHG', 'N2O', 'CO2', 'CH4'], axis=1)
inventory_2022 = inventory_2000_2021
inventory_2022['2022'] = inventory_2000_2021['2021']
inventory_2022
```

# Livestock

```python
"""

This code is for Livestock only in the agriculture dataset

"""
livestock = agriculture_census[(agriculture_census['Sublabel'] == 'Other Livestock') | (agriculture_census['Sublabel'] == 'Cattle') | (agriculture_census['Sublabel'] == 'Swine') | (agriculture_census['Sublabel'] == 'Poultry')]
livestock = livestock[livestock['Domain'] == 'TOTAL']
livestock = livestock[livestock['Data Item'].str.contains('MEASURED IN HEAD')]
livestock_cleaned = livestock[['County', 'State', 'Year', 'Data Item', 'Value', 'Sublabel']]
livestock_cleaned = livestock_cleaned[~livestock_cleaned['Value'].str.contains('D')]
livestock_cleaned['Value'] = livestock_cleaned['Value'].str.replace(',', '').astype(int)
livestock_dict = {'Cattle': ['CATTLE, INCL CALVES'], 'Swine': ['HOGS - SALES'], 'Poultry': ['POULTRY TOTALS - HATCHED'], 'Other Livestock': ['EQUINE, HORSES & PONIES - SALES', 'GOATS - SALES', 'SHEEP, INCL LAMBS - SALES']}

#Creating subtable for cattle only (exclude Bison)
def livestock_emissions(livestock_type):
  animal = livestock_cleaned[livestock_cleaned['Sublabel'] == livestock_type].sort_values(by = 'Value', ascending = False)
  animal = animal[animal['Data Item'].str.contains('|'.join(livestock_dict[livestock_type]))]
  animal.drop('Sublabel', axis = 1, inplace = True)
  animal.columns = [x.lower() for x in animal.columns]
  animal = animal.groupby(['county', 'year'],as_index=False).agg(sum)
  animal = animal.pivot(columns = 'year', index = 'county', values = 'value').fillna(0)
  animal_proportions = animal.div(animal.sum(axis=0), axis=1)
  combined_df = pd.DataFrame()
  for col in animal_proportions.columns:
    if (col != 1997):
      combined_df[col] = animal_proportions[col] * inventory_2000_2021.loc[livestock_type, str(col)]
  for county in county_list:
    if county not in combined_df.index:
      combined_df.loc[county] = [0, 0, 0, 0, 0]
  return combined_df.sort_index()

#example code here below will create a county-level table for 2002 - 2022 for other livestock (horse, sheep, and goats)
livestock_emissions('Other Livestock')

#For NaN values in the pivot, we replaced the NaN values with 0
```

# Crops

```python
#create new dataframe to filter needed columns
agriculture_cleaned2 = agriculture_census[['County', 'State', 'Year', 'Data Item', 'Value']]
#filter for only agriculture yield rows
agriculture_cleaned2 = agriculture_cleaned2[agriculture_cleaned2['Data Item'].str.contains(' - PRODUCTION, ')]
#split data item column to open units of measurement
units = agriculture_cleaned2['Data Item'].str.split(' - PRODUCTION, MEASURED IN')
# #Add new column called units containing the units, reassign data units to simply names of agricultural item, and rename columns
agriculture_cleaned2['units'] = units.apply(lambda x: x[1])
agriculture_cleaned2['Data Item'] = units.apply(lambda x: x[0])
agriculture_cleaned2.rename(columns={'Data Item':'crop_or_livestock','Value':'yield','County':'county','State':'state','Year':'year'}, inplace=True)
#Filter out the non integer values in yield
agriculture_cleaned2 = agriculture_cleaned2[agriculture_cleaned2['yield'] != ' (D)']
#Convert yield values into int
agriculture_cleaned2['yield'] = agriculture_cleaned2['yield'].str.replace(",", "").astype(int)
agriculture_cleaned2
```

```python
#create new dataframe to filter needed columns
agriculture_cleaned2 = agriculture_census[['County', 'State', 'Year', 'Commodity','Data Item', 'Value']].copy()
#filter for only agriculture yield rows measured in acres
agriculture_cleaned2 = agriculture_cleaned2[agriculture_cleaned2['Data Item'].str.contains(' - ACRES')]
#split data item column to open units of measurement
units = agriculture_cleaned2['Data Item'].str.split(' - ACRES')
# #Add new column called units containing the units, reassign data units to simply names of agricultural item, and rename columns
agriculture_cleaned2['units'] = units.apply(lambda x: x[1])
agriculture_cleaned2['Data Item'] = units.apply(lambda x: x[0])
agriculture_cleaned2.rename(columns={'Data Item':'crop_or_livestock','Value':'yield','County':'county','State':'state','Year':'year'}, inplace=True)
#Filter out the non integer values in yield
agriculture_cleaned2 = agriculture_cleaned2[agriculture_cleaned2['yield'] != ' (D)']
agriculture_cleaned2 = agriculture_cleaned2[agriculture_cleaned2['yield'] != ' (Z)']
#Filter out all subcategories by taking the max of each commodity for each county for each year
agriculture_cleaned2 = agriculture_cleaned2.groupby(['county','year','Commodity'], as_index=False)['yield'].max()
#Convert yield values into int
agriculture_cleaned2['yield'] = agriculture_cleaned2['yield'].str.replace(',', '').astype(int)
agriculture_cleaned2
```

```python
nitrogen_application = pd.read_csv('/Nitrogen Application.csv')
```

```python
nitrogen_application
```

```python
# Fertilizer
crops_fertilizer = agriculture_cleaned2.merge(nitrogen_application, on='Commodity')
crops_fertilizer['Fertilizer (lbs)'] = crops_fertilizer['Fertilizer Needed (lbs/acre)'] * crops_fertilizer['yield']
crops_fertilizer = crops_fertilizer.groupby(['county', 'year'], as_index=False)['Fertilizer (lbs)'].sum()
crops_fertilizer = crops_fertilizer.pivot(columns = 'year', index = 'county', values = 'Fertilizer (lbs)').fillna(0)
crops_proportions = crops_fertilizer.div(crops_fertilizer.sum(axis=0), axis=1)
combined_df = pd.DataFrame()
for col in crops_proportions.columns:
    if (col != 1997):
      combined_df[col] = crops_proportions[col] * inventory_2000_2021.loc['Crop Fertilizer', str(col)]
for county in county_list:
  if county not in combined_df.index:
    combined_df.loc[county] = [0, 0, 0, 0, 0]
combined_df
```

```python
# Rice
rice_production = agriculture_cleaned2[(agriculture_cleaned2['Commodity'] == 'RICE') | (agriculture_cleaned2['Commodity'] == 'WILD RICE')]
rice_production = rice_production.groupby(['county', 'year'], as_index=False)['yield'].agg(sum)
rice_production = rice_production.pivot(columns = 'year', index = 'county', values = 'yield').fillna(0)
rice_production = rice_production.div(rice_production.sum(axis=0), axis=1)
rice_emissions = pd.DataFrame()
for col in rice_production.columns:
    if (col != 1997):
      rice_emissions[col] = rice_production[col] * inventory_2000_2021.loc['Rice Emissions', str(col)]
for county in county_list:
  if county not in rice_emissions.index:
    rice_emissions.loc[county] = [0, 0, 0, 0, 0]
rice_emissions
```

```python
# Burning - Rice
rice_burning = pd.DataFrame()
for col in rice_production.columns:
    if (col != 1997):
      rice_burning[col] = rice_production[col] * inventory_2000_2021.loc['Crop Burning - Rice', str(col)]
for county in county_list:
  if county not in rice_burning.index:
    rice_burning.loc[county] = [0, 0, 0, 0, 0]

# Burning - Barley
barley_production = agriculture_cleaned2[(agriculture_cleaned2['Commodity'] == 'BARLEY')]
barley_production = barley_production.groupby(['county', 'year'], as_index=False)['yield'].agg(sum)
barley_production = barley_production.pivot(columns = 'year', index = 'county', values = 'yield').fillna(0)
barley_production = barley_production.div(barley_production.sum(axis=0), axis=1)
barley_burning = pd.DataFrame()
for col in barley_production.columns:
    if (col != 1997):
      barley_burning[col] = barley_production[col] * inventory_2000_2021.loc['Crop Burning - Barley', str(col)]
for county in county_list:
  if county not in barley_burning.index:
    barley_burning.loc[county] = [0, 0, 0, 0, 0]

# Burning - Wheat
wheat_production = agriculture_cleaned2[(agriculture_cleaned2['Commodity'] == 'WHEAT')]
wheat_production = wheat_production.groupby(['county', 'year'], as_index=False)['yield'].agg(sum)
wheat_production = wheat_production.pivot(columns = 'year', index = 'county', values = 'yield').fillna(0)
wheat_production = wheat_production.div(wheat_production.sum(axis=0), axis=1)
wheat_burning = pd.DataFrame()
for col in wheat_production.columns:
    if (col != 1997):
      wheat_burning[col] = wheat_production[col] * inventory_2000_2021.loc['Crop Burning - Wheat', str(col)]
for county in county_list:
  if county not in wheat_burning.index:
    wheat_burning.loc[county] = [0, 0, 0, 0, 0]

# Burning - Almonds
almond_production = agriculture_cleaned2[(agriculture_cleaned2['Commodity'] == 'ALMONDS')]
almond_production = almond_production.groupby(['county', 'year'], as_index=False)['yield'].agg(sum)
almond_production = almond_production.pivot(columns = 'year', index = 'county', values = 'yield').fillna(0)
almond_production = almond_production.div(almond_production.sum(axis=0), axis=1)
almond_burning = pd.DataFrame()
for col in almond_production.columns:
    if (col != 1997):
      almond_burning[col] = almond_production[col] * inventory_2000_2021.loc['Crop Burning - Almond', str(col)]
for county in county_list:
  if county not in almond_burning.index:
    almond_burning.loc[county] = [0, 0, 0, 0, 0]

# Burning - Walnuts
walnut_production = agriculture_cleaned2[(agriculture_cleaned2['Commodity'] == 'WALNUTS')]
walnut_production = walnut_production.groupby(['county', 'year'], as_index=False)['yield'].agg(sum)
walnut_production = walnut_production.pivot(columns = 'year', index = 'county', values = 'yield').fillna(0)
walnut_production = walnut_production.div(walnut_production.sum(axis=0), axis=1)
walnut_burning = pd.DataFrame()
for col in walnut_production.columns:
    if (col != 1997):
      walnut_burning[col] = walnut_production[col] * inventory_2000_2021.loc['Crop Burning - Walnut', str(col)]
for county in county_list:
  if county not in walnut_burning.index:
    walnut_burning.loc[county] = [0, 0, 0, 0, 0]

total_burning = almond_burning + walnut_burning + wheat_burning + barley_burning + rice_burning
total_burning
```
